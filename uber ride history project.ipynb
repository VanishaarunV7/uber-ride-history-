```python
plt.figure(figsize=(7,7))
def autopct_with_counts(pct, values):
    total = sum(values)
    absolute = int(pct / 100.*total)
    return f'{pct:.1f}% ( {absolute} )'
    
uber_df['CATEGORY*'].value_counts().plot.pie(autopct = lambda pct: autopct_with_counts(pct, uber_df['CATEGORY*'].value_counts()), startangle = 90, cmap = 'Set3', labels = None)
plt.title('Proportion of Categories with Counts')
plt.ylabel('')
plt.legend(labels = uber_df['CATEGORY*'].value_counts().index, loc = 'upper right')
plt.show()
```
